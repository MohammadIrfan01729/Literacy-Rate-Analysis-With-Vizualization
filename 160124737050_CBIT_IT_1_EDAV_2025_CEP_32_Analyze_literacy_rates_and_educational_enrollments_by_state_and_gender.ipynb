# Data Display


```python
print(df.head(),"\n")
print(df.tail())
```

# Average Literacy Rate

```python
# Import the required Libraries
import pandas as pd
import numpy as np

# Average Of Literacy Rates State Wise Yearly
avg_literacy = df.groupby(['State/Region', 'Gender'])['Literacy Rate'].mean()
print(avg_literacy)
```

# Filter Data By Gender


```python
# Filter by Male
df_male = df[df['Gender'] == 'Male']

# Filter by Female
df_female = df[df['Gender'] == 'Female']

# Filtered Data By Gender
print('Filtered By Male:\n')
print(df_male.head(),"\n")
print('Filtered By Female:\n')
print(df_female.head())
```

# Handiling Missing Enrollments


```python
# Filling The Missing The Enrollments With Mean Enrollment value
mean_value = df['Enrollments'].mean()
df.fillna({'Enrollments': mean_value}, inplace=True)
print(df.head())
```

# Group Literacy Stats By Age & State


```python
# Group Literacy Stats By Age & State/Region using groupby()
Group_Stats = df.groupby(['State/Region', 'Age'])['Literacy Rate'].mean()
print(Group_Stats)
```

# Gender - wise comparision plot

```python
import matplotlib.pyplot as plt

# Groupint the states and gender with Average Literacy rates
Avg_df = df.groupby(['State/Region', 'Gender'])['Literacy Rate'].mean().reset_index()
print(Avg_df,'\n')

# Pivot the DataFrame for better plotting of Gender Wise Comparision
Pivot_df = Avg_df.pivot(index='State/Region', columns='Gender', values='Literacy Rate')
print(Pivot_df)

# Plot Code
Pivot_df.plot(kind='bar', figsize=(14, 6))
plt.title('Gender-wise Literacy Rate Comparison by State')
plt.ylabel('Average Literacy Rate (%)')
plt.xlabel('State/Region')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```
